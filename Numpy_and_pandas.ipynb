```python
def binary_addition_decimal(binary1, binary2):
  """
  Adds two binary numbers by converting to decimal, adding, and converting back.

  Args:
    binary1: A string representing the first binary number.
    binary2: A string representing the second binary number.

  Returns:
    A string representing the sum of the two binary numbers.
  """
  # Convert binary strings to integers
  decimal1 = int(binary1, 2)
  decimal2 = int(binary2, 2)

  # Add the decimal numbers
  decimal_sum = decimal1 + decimal2

  # Convert the sum back to a binary string (remove the "0b" prefix)
  binary_sum = bin(decimal_sum)[2:]

  return binary_sum

# Example usage:
binary_num1 = "11" # 3 in decimal
binary_num2 = "1"  # 1 in decimal
result = binary_addition_decimal(binary_num1, binary_num2)
print(f"The sum of {binary_num1} and {binary_num2} is: {result}")

binary_num3 = "1010" # 10 in decimal
binary_num4 = "1011" # 11 in decimal
result2 = binary_addition_decimal(binary_num3, binary_num4)
print(f"The sum of {binary_num3} and {binary_num4} is: {result2}")
```

### Selecting Rows
You can select rows by their integer location using `iloc[]` or by their label (index) using `loc[]`.

Using `iloc[]` to select rows by integer position:

```python
# Select the first row using iloc
first_row = data.iloc[0]
display(first_row)

# Select the first 5 rows using iloc
first_five_rows = data.iloc[0:5]
display(first_five_rows)

# Select rows with specific integer positions
specific_rows_iloc = data.iloc[[0, 2, 4]]
display(specific_rows_iloc)
```

Using `loc[]` to select rows by index label. In this DataFrame, the index is the default integer index, so `loc[]` works similarly to `iloc[]` for basic integer indexing, but it's designed for label-based indexing.

```python
# Select the row with index label 0 using loc
row_by_label = data.loc[0]
display(row_by_label)

# Select rows with index labels from 0 to 4 (inclusive) using loc
rows_by_label_range = data.loc[0:4]
display(rows_by_label_range)

# Select rows with specific index labels
specific_rows_loc = data.loc[[0, 2, 4]]
display(specific_rows_loc)
```

You can also combine row and column selection using `iloc[]` and `loc[]`.
